## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

Which type of Hotel is most preffered by the guests?

```python
# Chart - 1 visualizsing by the pie chart
df['hotel'].value_counts().plot.pie(explode=[0.05,0.05],autopct='%1.1f%%',shadow=True,figsize=(5,7),fontsize=20)
plt.title('Hotel Booking Percentage')
plt.axis('equal')
plt.show()
```

**Observation**
City Hotel is most preffered hotel by guests. Thus City Hotel has maximum bookings.

2) Which Agent made  the most bookings?

```python
highest_bookings = df.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count':'Most_Booking'}).sort_values(by='Most_Booking',ascending=False)
top_ten_highest_bookings = highest_bookings[:10]
highest_bookings
```

#### Chart - 2

```python
# Chart - 2 visualizaing the graph.

plt.figure(figsize=(10,5))
sns.barplot(x=top_ten_highest_bookings['agent'],y=top_ten_highest_bookings['Most_Booking'],order=top_ten_highest_bookings['agent'])
plt.xlabel('Agent no')
plt.ylabel('Number of Bookings')
plt.title('Most Booking by the Agent')
plt.show()
```

Observation
Agent ID no 9 made most of the bookings.
